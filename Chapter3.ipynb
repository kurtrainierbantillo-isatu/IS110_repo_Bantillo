#### Exercise 2

Write your salary computation to give an employee 1.5 times the hourly rate for hours worked above 40 hours. Example:

    Enter Hours: 45
    Enter Rate: 10
    Pay: 475.0
    
[PEW] E3.2

```python
Hours=input('Enter Hours: ')
HourlyRate=input('Enter Hourly Rate: ')
Hours=float(Hours)
HourlyRate=float(HourlyRate)

if Hours > 40:
    Pay=40*HourlyRate+(Hours-40)*HourlyRate*1.5
else:
    Pay=Hours*HourlyRate

print('Pay:', Pay)
```

## CHALLENGE

Write a program to prompt for a score between 0 and 100 and print a grade using the following table:

    Score           Grade
    [ 98, 100]      1.0
    [ 94,  97)		1.25
    [ 90,  93)		1.5
    [ 86,  89)		1.75
    [ 82,  85)		2.0
    [ 78,  81)		2.25
    [ 75,  77)		2.5
    [ 0,   74)		5.0

Adapted from [PEW] E3.3
<br>

```python
score = input("Enter Score: ")
scorenum = float(score)

if scorenum >= 98 and scorenum <= 100:
    print("Your Grade is 1.0")
elif scorenum >= 94 and scorenum <= 97:
    print("Your Grade is 1.25")
elif scorenum >= 93 and scorenum <= 90:
    print("Your Grade is 1.5")
elif scorenum >= 86 and scorenum <= 89:
    print("Your Grade is 1.75")
elif scorenum >= 82 and scorenum <= 85:
    print("Your Grade is 2.0")
elif scorenum >= 78 and scorenum <= 81:
    print("Your Grade is 2.25")
elif scorenum >= 75 and scorenum <= 77:
    print("Your Grade is 2.5")
elif scorenum >= 0 and scorenum <= 74:
    print("Your Grade is 5.0")
else :
    print('Invalid')
```
