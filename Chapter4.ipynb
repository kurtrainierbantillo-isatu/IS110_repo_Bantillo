## CHALLENGE

Write a function called computegrade that takes a score as its parameter and returns a grade as a string using the following table:

    Score           Grade
    [ 98, 100]      1.0
    [ 94,  97)		1.25
    [ 90,  93)		1.5
    [ 86,  89)		1.75
    [ 82,  85)		2.0
    [ 78,  81)		2.25
    [ 75,  77)		2.5
    [ 0,   74)		5.0

    Run the program repeatedly to test the various different values for input.
    
Adapted from [PEW] E4.6
<br>

```python
def computegrade(score):
    if score >= 98 and score <= 100:
        return "1.0"
    elif score >= 94 and score < 98:
        return "1.25"
    elif score >= 90 and score < 94:
        return "1.5"
    elif score >= 86 and score < 90:
        return "1.75"
    elif score >= 82 and score < 86:
        return "2.0"
    elif score >= 78 and score < 82:
        return "2.25"
    elif score >= 75 and score < 78:
        return "2.5"
    elif score >= 0 and score < 75:
        return "5.0"
    else:
        return "Invalid Score"

score_input = input("Enter Score: ")
try:
    score = float(score_input)
    grade = computegrade(score)
    print("Your Grade is", grade)
except ValueError:
    print("Invalid input. Please enter a numerical score.")
```
